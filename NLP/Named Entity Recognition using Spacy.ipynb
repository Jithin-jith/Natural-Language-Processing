# Named Entity Recognition

Named Entity Recognition (NER) is a subtask of information extraction that seeks to locate and classify named entities mentioned in unstructured text into predefined categories. 

These categories typically include entities such as:

**Persons:** Names of people (e.g., "Barack Obama").
<br>
**Organizations:** Names of companies, institutions, etc. (e.g., "Google").
<br>
**Locations:** Geographical places (e.g., "Paris").
<br>
**Dates and Times:** Specific dates and times (e.g., "January 1, 2020").
<br>
**Monetary Values:** Amounts of money (e.g., "$100").
<br>
**Percentages:** Percent values (e.g., "50%").
<br>
**Other Entities:** Depending on the application, other categories such as products, events, works of art, etc.

```python
#Importing libraries and models

import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

#### About spaCy


spaCy is an open-source software library for advanced natural language processing (NLP) in Python. It's designed specifically for production use and provides a fast and efficient way to process and analyze large volumes of text. spaCy offers pre-trained models for various languages and includes tools for **tokenization**, **part-of-speech tagging**, **named entity recognition**, **dependency parsing**, and more.

#### Models in spaCy

spaCy offers several types of models, each designed for specific tasks and language requirements. These models vary in size and capabilities, and they include:

- **Small Models:** Lightweight and fast, suitable for tasks where speed is critical and memory is limited (e.g., en_core_web_sm).
- **Medium Models:** Balanced between speed and accuracy, providing better performance for most tasks (e.g., en_core_web_md).
- **Large Models:** Highly accurate but require more memory and processing power (e.g., en_core_web_lg).
- **Transformer Models:** Based on transformer architectures (like BERT), these models offer state-of-the-art accuracy (e.g., en_core_web_trf).

```python
#lets look at the components of the pipelines in the loaded model\
nlp.pipe_names
```

These are the components of the pipelines in the above "en_core_web_sm" model. The components in the pipeline are applied sequentially, one after the other, in the order they appear in pipe_names. This means that each component takes the output of the previous component as its input.

#### Example of Pipeline working

```python
#1.Load the Model and Process the Text

# Load the small English model
nlp = spacy.load("en_core_web_sm")

# Example text
text = "Apple is looking at buying U.K. startup for $1 billion."

# Process the text
doc = nlp(text)
```

```python
#2. Tokenization
#Tokenization is the first step where the text is split into individual tokens.

print("Tokens:")
for token in doc:
    print(token.text)
```

```python
#3.Token-to-Vector (tok2vec)
#Converts each token into a dense vector representation.

print("\nToken Vectors (First 3 Dimensions):")
for token in doc:
    print(f"{token.text}: {token.vector[:3]}")
```

```python
#4. Part-of-Speech Tagging (POS)
#The POS tagger assigns part-of-speech tags to each token.

print("POS Tags:\n")
for token in doc:
    print(f"{token.text}: {token.pos_}")
```

```python
#5. Dependency Parsing
#The dependency parser identifies the syntactic relationships between tokens.

print("Dependency Parsing:\n")
for token in doc:
    print(f"{token.text}: {token.dep_} (Head: {token.head.text})")
```

```python
#6. Attribute Ruler and lemmatization
#The attribute ruler modifies or sets attributes on tokens according to predefined rules.

print("Attribute Ruler (Lemmas):\n")
for token in doc:
    print(f"{token.text}: {token.lemma_}")
```

```python
#7. Named Entity Recognition (NER)
#Identifies and classifies named entities in the text.

print("Named Entities:\n")
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```

#### Correcting Misclassification of Entities

In the above example we can see that AI and IoT are misclassified into ORG category. Let's assume that we want to classify them into a new category called 'Technology' which is not already present in the ner categories of the above model.

```python
#load the model again
nlp = spacy.load("en_core_web_sm")

# Create an EntityRuler
ruler = nlp.add_pipe("entity_ruler", before="ner")

# Define patterns for the custom entities
patterns = [
    {"label": "TECHNOLOGY", "pattern": "AI"},
    {"label": "TECHNOLOGY", "pattern": "IoT"}
]

# Add patterns to the ruler
ruler.add_patterns(patterns)
```

```python
# Process the text
doc = nlp(text)

# Print named entities
print("Named Entities:\n")
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```
